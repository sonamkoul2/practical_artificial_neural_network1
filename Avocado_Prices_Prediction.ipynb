```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
from fbprophet import Prophet
```

```python
# load csv file data as a dataframe
avocado_df = pd.read_csv(r"C:\Users\sonam\OneDrive\Desktop\DeepLearningPacticalWorkout\DL and ML Practical Tutorials - Package\Project 4\avocado.csv")
```

```python
# The first five data
avocado_df.head()
```

```python
# Observe the last few raws
avocado_df.tail()
```

```python
# sort the dataframe by date
avocado_df = avocado_df.sort_values("Date")
```

```python
# visualize 
plt.figure(figsize=(10, 10))
plt.plot(avocado_df["Date"], avocado_df["AveragePrice"])
```

```python
avocado_df
```

```python
plt.figure(figsize=[25, 12])
sns.countplot(x = 'region', data=avocado_df)
plt.xticks(rotation=45)
```

```python
plt.figure(figsize=[25, 10])
sns.countplot(x = 'year', data=avocado_df)
plt.xticks(rotation=45)
```

```python
avocado_prophet_df = avocado_df[["Date", "AveragePrice"]]
```

```python
avocado_prophet_df = avocado_prophet_df.rename(columns={"Date":'ds', 'AveragePrice': 'y'})
```

```python
avocado_prophet_df
```

```python
m = Prophet()
m.fit(avocado_prophet_df)
```

```python
future = m.make_future_dataframe(periods=365)
forecast = m.predict(future)
```

```python
forecast
```

```python
figure = m.plot(forecast, xlabel='Date', ylabel='Price')
```

```python
figure3 = m.plot_components(forecast)
```
